Create the leaflet widget to display the map

```python
from leafletwidget import (
    Map,
    Marker,
    TileLayer, ImageOverlay,
    Polyline, Polygon, Rectangle, Circle, CircleMarker,
    GeoJSON,
    DrawControl
)
from leafletwidget import initialize_notebook

from IPython.html import widgets
from IPython.display import display

initialize_notebook()
```

calculate center of the map and create map variable (m)

```python
lat=[]
lon=[]
for gj in [zones,stations]:
    for f in zones['features']:
        for c in f['geometry']['coordinates'][0]: #0 because had to embed coordinates list inside of another list
            lon.append(c[0])
            lat.append(c[1]) 

center = [(max(lat)+min(lat))/2, (max(lon) + min(lon))/2]
m = Map(width='100%',height='600px', center=center, zoom=9)
```

Add zones and stations to map, then display

```python
gjzones = GeoJSON(data=zones)
m.add_layer(gjzones)

gjstations = GeoJSON(data=stations)
m.add_layer(gjstations)

display(m)
```

```python
m.remove_layer(gjzones)
m.remove_layer(gjstations)
```
